```python
import numpy as np
import matplotlib.pyplot as plt

#Generar nuestros vectores
x = np.array([[0,0], [0,1], [1,0], [1,1]]) #variables independientes  #entradas  -> 2+2
y = np.array([0,1,1,0]) #variables dependientes                     #target    -> 4

#Grafico
plt.scatter(x[:,0], x[:,1], c=y, cmap='winter')
plt.xlabel('característica 1')
plt.ylabel('característica 2')
plt.title('Problema XOR con SVM')
plt.grid()
plt.show()
```

## Implementación del modelo SVM

```python
from sklearn.svm import SVC

modelo = SVC(kernel='linear') #creado el modelo
modelo.fit(x,y) # ajustar el modelo

y_estimada = modelo.predict(x) #predicción
print(f'Y estimada por el modelo: {y_estimada}')
print(f'Y real: {y}')
```

```python
#Frontera de decisión
x_min, x_max = x[:,0].min()-1, x[:,0].max()+1
y_min, y_max = x[:,1].min()-1, x[:,1].max()+1

xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))

#Frontera con el gráfico
y_graf = modelo.predict(np.c_[xx.ravel(), yy.ravel()])
y_graf = y_graf.reshape(xx.shape)

plt.contourf(xx, yy, y_graf, cmap='winter', alpha=0.5)
plt.scatter(x[:,0], x[:,1], c=y, cmap='winter')
plt.xlabel('característica 1')
plt.ylabel('característica 2')
plt.title('Problema XOR con SVM')
plt.grid()
plt.show()
```

## Probar con un nuevo punto

```python
x_test = np.array([[1.0, -0.5]])

y_test_nuevo_valor = modelo.predict(x_test)
print(f'Valor estimado para el nuevo punto: {y_test_nuevo_valor}')

# Probar con una frontera límite
x_test = np.array([[0.5, 0.5]])

y_test_nuevo_valor = modelo.predict(x_test)
print(f'Valor estimado para el nuevo punto: {y_test_nuevo_valor}')
```
